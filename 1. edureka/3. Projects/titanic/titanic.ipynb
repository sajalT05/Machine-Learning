#### Data Wrangling: cleaning null values and unnecessary data from the dataset

```python
dataset.head(10)
```

```python
# remove cabin column
dataset.drop("Cabin",axis=1,inplace=True)
```

```python
dataset.head(5)
```

```python
# drop rows with null values
dataset.dropna(inplace=True)
```

```python
dataset.head(10)
```

```python
sns.heatmap(dataset.isnull(),yticklabels=False,cbar=False)
```

```python
dataset.isnull().sum()
```

```python
dataset.head(10)
```

create new datasets with descrete values from a cloumn

```python
# passenger class
# drop first column
passengerClass=pd.get_dummies(dataset['Pclass'],drop_first=True)
passengerClass.head(5)
```

```python
# sex
passengerSex=pd.get_dummies(dataset['Sex'],drop_first=True)
passengerSex.head(5)
```

```python
# embarked
passengerEmbarked=pd.get_dummies(dataset['Embarked'],drop_first=True)
passengerEmbarked.head(5)
```

```python
# add new descrete data column in main table
dataset=pd.concat([dataset,passengerClass,passengerSex,passengerEmbarked],axis=1)
dataset.head(5)
```

```python
dataset.info()
```

```python
# drop columns converted into descrete columns
dataset.drop(['PassengerId','Pclass','Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```

```python
dataset.head(5)
```

#### Train Data

```python
X=dataset.drop("Survived",axis=1)
y=dataset["Survived"]
```

```python
# split data into training and testing phase
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```
